```python
# Regular Expression Codes

START_OF_LINE = r"^"
OPTIONAL = "?"
ANYTHING = "."
ZERO_OR_MORE = "*"
ONE_OR_MORE = "+"

SPACE = "\s"
SPACES = SPACE + ONE_OR_MORE
NOT_SPACE = "[^\s]" + ONE_OR_MORE
EVERYTHING_OR_NOTHING = ANYTHING + ZERO_OR_MORE

ERASE = ""
FORWARD_SLASH = "\/"
NEWLINES = r"[\r\n]"

#Twitter Specific Codes
HYPERLINKS = ("http" + "s" + OPTIONAL + ":" + FORWARD_SLASH + FORWARD_SLASH
              + NOT_SPACE + NEWLINES + ZERO_OR_MORE)
RE_TWEET = START_OF_LINE + "RT" + SPACES
HASH = "#"
AT = "(?:\@)\S+"

def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)

def rm_symbols(text):
  no_link = re.sub(HYPERLINKS,ERASE,text)
  no_hash = re.sub(HASH + r'\w+', ERASE, no_link)
  no_hash2 = re.sub(HASH, ERASE, no_hash)
  no_retweet = re.sub(RE_TWEET,ERASE,no_hash2)
  no_at = re.sub(AT,ERASE,no_retweet)
  no_newline = re.sub("\n",ERASE,no_at)
  return no_newline
```

```python
#Removing Emojies (Dylan)
try:
  data["new_text"] = data["text"].progress_apply(remove_emojis)
except:
  print("Removing emoji failed")
  exit()
  
#Removing Symbols (Dylan)
try:
  data["new_text"] = data["new_text"].progress_apply(rm_symbols)
except:
  print("Removing symbols failed")
  exit()



```
